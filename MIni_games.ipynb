### 1. Matching game: Using two lists and simple codes, you can create a variety of fun such as a quiz, a spelling game and others.

```python
question = ['What is the bigger ocean on earth?', 'Who is the nasty boss of Meta?', 'What is 17% of 84?', 'What does CO2 stands for?']
answer = ['Pacific ocean', 'Mark Zuckerberg', '14.28', 'Carbon dioxide']

n = len(question) # Counting how many items in the question list
right = 0 # Accumulate the right items
for i in range(n): # loop the questions and matching them with the answers in the answer list
  y = input(f'{i + 1}. {question[i]}\n')
  normalised_y = y.lower()
  if normalised_y == answer[i].lower(): # Using lower() to ignore case sensitive issues in both input and answer fields
    print('You got it!')
    right += 1
  else:
    print(f'Sorry, wrong answer! The correct answer is {answer[i]}.')

print('You got', right, 'right! \U0001F680')
print('And you got', n-right, 'wrong!')
```

### 2. Rock, Paper, Scissors: This game applies the 'random' and the if..else is good for logical thinking.

```python
import random
hand = ['paper', 'scissor', 'rock']
you_win = 0
comp_win = 0

for i in range(5):
  q1 = int(input('Let\'s play! 0=Paper, 1=Scissor, 2=Rock\n'))
  q2 = random.randint(0,2) # Using 'random.randint' to generate an integer between 0 and 2

  if q1 == q2:
    print(f'Yours is {hand[q1]} and computer is {hand[q2]}. Equal!') # Matching the number with the 'hand' list to show the name of choice.
  elif q1 == 0:
    if q2 == 1:
      print(f'Yours is {hand[q1]} and computer is {hand[q2]}. Sorry! you lost!')
      comp_win += 1
    else:
      print(f'Yours is {hand[q1]} and computer is {hand[q2]}. You won!')
      you_win += 1
  elif q1 == 1:
    if q2 == 0:
      print(f'Yours is {hand[q1]} and computer is {hand[q2]}. You won!')
      you_win += 1
    else:
      print(f'Yours is {hand[q1]} and computer is {hand[q2]}. Sorry! you lost!')
      comp_win += 1
  elif q1 == 2:
    if q2 == 0:
      print(f'Yours is {hand[q1]} and computer is {hand[q2]}. Sorry! you lost!')
      comp_win += 1
    else:
      print(f'Yours is {hand[q1]} and computer is {hand[q2]}. You won!')
      you_win += 1
  else: # This one copes with the wrong number input which is not between 0 and 2
    print('Wrong input! You lost one point!')
    comp_win += 1

if you_win == comp_win:
  print('Even!')
elif you_win > comp_win:
  print('You got', you_win, '! You won! \U0001F44F')
else:
  print('You lost! Computer bit you with', comp_win, 'right!')
```
